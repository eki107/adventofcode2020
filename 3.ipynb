```python
def move_through_slope(slope, right):
    x = 0
    path = ""
    
    for y in slope.split():
        path += y[x] # mark path

        x = (x + right) % len(y) # move and wraparound
    
    return path
```

```python
res = [
 move_through_slope(pinp, 1).count("#"),
 move_through_slope(pinp, 3).count("#"), # answer for 1st
 move_through_slope(pinp, 5).count("#"),
 move_through_slope(pinp, 7).count("#"),
 move_through_slope("\n".join(pinp.split()[::2]), 1).count("#"), # same rules, only every other row
]
```

```python
from functools import reduce
res[1], reduce(lambda x, y: x * y, res)
```
