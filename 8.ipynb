```python
# split rows and first 3 letters of every row is an instruction and everything after 4th letter is an arg
instructions = [(i[:3], int(i[4:])) for i in pinp.split("\n")]
```

```python
def execute_program(instructions):
    i, acc, k = 0, 0, 0 # instruction pointer, accumulator, instruction counter
    instruction_counter = len(instructions) * [0] # instruction list order
    code_run = True # success
    
    # my code break will be when there is an infinite loop or end of instruction list
    while True:
        
        # first, test 
        if i >= len(instructions):
            break
        elif instruction_counter[i] > 0:
            code_run = False
            break
        else:
            instruction_counter[i] = k

        # execute instruction
        cur_instruction = instructions[i]    
        
        if cur_instruction[0] == "nop":
            i += 1
        elif cur_instruction[0] == "acc":
            acc += cur_instruction[1]
            i += 1
        elif cur_instruction[0] == "jmp":
            i += cur_instruction[1]

        k += 1
    
    # return accumulator of run, instruction list with order, 
    # succes/failure of code run, number of instructions successfully executed
    return acc, list(zip(instructions, instruction_counter)), code_run, k
```
